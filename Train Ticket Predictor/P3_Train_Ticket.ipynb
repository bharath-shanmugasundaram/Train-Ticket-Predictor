```python
Temp=df.to_numpy(int)
Y=Temp[:,-1]
X=Temp[:,:-1]
X
Y = Y.reshape(-1, 1)   
Y
```

## Initializing Parameter

```python
np.random.seed(42)
W1 = np.random.randn(13, X.shape[1])   
b1 = np.zeros((1, 13))                 

W2 = np.random.randn(10, 13)            
b2 = np.zeros((1, 10))                 

W3 = np.random.randn(7, 10)            
b3 = np.zeros((1, 7))    

W4 = np.random.randn(3, 7)            
b4 = np.zeros((1, 3))  

W5 = np.random.randn(1, 3)            
b5 = np.zeros((1, 1))  

learning_rate = 0.1
epochs = 10000
batch_size = 64 
```

```python
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

def sigmoid_derivative(a):
    return a * (1 - a)
```

```python
cost=[]
for epoch in range(epochs):

    permutation = np.random.permutation(X.shape[0])
    X_shuffled = X[permutation]
    Y_shuffled = Y[permutation]

    epoch_loss = 0
    num_batches = X.shape[0] // batch_size
    for i in range(0, X.shape[0], batch_size):
    # Forward pass
        Z1 = np.dot(X, W1.T) + b1         
        A1 = sigmoid(Z1)

        Z2 = np.dot(A1, W2.T) + b2        
        A2 = sigmoid(Z2)

        Z3= np.dot(A2,W3.T) + b3
        A3= sigmoid(Z3)

        Z4= np.dot(A3,W4.T) + b4
        A4= sigmoid(Z4)

        Z5 = np.dot(A4,W5.T) + b5
        A5 = sigmoid(Z5)
        

        m = X.shape[0]
        loss = -(1/m) * np.sum(Y*np.log(A5+1e-8) + (1-Y)*np.log(1-A5+1e-8))

        # Backward pass

        dZ5 = A5 - Y                         
        dW5 = (1/m) * np.dot(dZ5.T, A4)     
        db5 = (1/m) * np.sum(dZ5, axis=0, keepdims=True) 

        dZ4 = np.dot(dZ5, W5)  * sigmoid_derivative(A4)  
        dW4 = (1/m) * np.dot(dZ4.T, A3)      
        db4 = (1/m) * np.sum(dZ4, axis=0, keepdims=True) 

        dZ3 = np.dot(dZ4, W4) * sigmoid_derivative(A3)  
        dW3 = (1/m) * np.dot(dZ3.T, A2)     
        db3 = (1/m) * np.sum(dZ3, axis=0, keepdims=True)

        dZ2 = np.dot(dZ3, W3) * sigmoid_derivative(A2)  
        dW2 = (1/m) * np.dot(dZ2.T, A1)      
        db2 = (1/m) * np.sum(dZ2, axis=0, keepdims=True)

        dZ1 = np.dot(dZ2, W2) * sigmoid_derivative(A1)  
        dW1 = (1/m) * np.dot(dZ1.T, X)    
        db1 = (1/m) * np.sum(dZ1, axis=0, keepdims=True)

        W1 -= learning_rate * dW1
        b1 -= learning_rate * db1
        W2 -= learning_rate * dW2
        b2 -= learning_rate * db2
        W3 -= learning_rate * dW3
        b3 -= learning_rate * db3
        W4 -= learning_rate * dW4
        b4 -= learning_rate * db4
        W5 -= learning_rate * dW5
        b5 -= learning_rate * db5
        
    cost.append(loss)
    print(f"Epoch {epoch+1}/{epochs} - Loss: {loss:.4f}")
```
